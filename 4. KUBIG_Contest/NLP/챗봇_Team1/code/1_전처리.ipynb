## cleansing 함수 정의

```python
def cleansing(df):
  temp=[]
  for i in range(0,len(df)):
    temp.append(df['review'][i])
  temp1 = ' '.join(s for s in temp)
  temp2 = re.sub('\n', ' ', temp1)
  temp3 = re.sub('\'', '', temp2)
  temp4 = re.sub('[^a-zA-z.]', ' ', temp3)
  words = temp4.lower().split()
  stops = set(stopwords.words('english'))
  meaningful_words = [w for w in words if not w in stops]
  temp5 = [wordnet_lemmatizer.lemmatize(w) for w in meaningful_words]
  return(' '.join(temp5))
```

### lemmatize vs stemming

```python
words=['strawberries.', 'strawberry', 'candies', 'candy', 'smell', 'smelt', 'comforting', 'comfort']
print([wordnet_lemmatizer.lemmatize(w) for w in words])
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer('english')
print([stemmer.stem(w) for w in words])
```
